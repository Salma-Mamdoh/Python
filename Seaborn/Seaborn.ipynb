```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flights=sns.load_dataset("flights")
flights
```

```python
feb_flights=flights[flights['month']=='Feb']
feb_flights
```

```python
sns.set_style("darkgrid")
sns.lineplot(x='year' , y='passengers' ,data=feb_flights )
plt.title("Year vs Passengers")
plt.show()
```

```python
sns.lineplot(x='year' , y='passengers' ,data=flights , hue="month")
plt.show()
```

```python
fm=sns.load_dataset('fmri')
fm
```

```python
sns.lineplot(data=fm , x="timepoint" , y="signal" , hue="event" , style="region")
plt.show()
```

## relplot

```python
sns.relplot(data=fm , x="timepoint" , y="signal" , hue="event" , style="region" , kind ="line", col="event")
plt.show()
```

```python
sns.relplot(data=fm , x="timepoint" , y="signal" , hue="event" , style="region" , kind ="line", col="region")
plt.show()
```

```python
sns.relplot(data=fm , x="timepoint" , y="signal" , hue="event" , style="region" , kind ="line", col="region",row="event")
plt.show()
```

## Scatter plots

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(data=tips , x="total_bill" , y="tip" , hue="sex")
plt.show()
```

```python
sns.scatterplot(data=tips , x="total_bill" , y="tip" , hue="sex" , style="smoker")
plt.show()
```

## Regression Plot

```python
sns.regplot(x="total_bill" , y="tip" , data=tips)
plt.show()
```

```python
sns.regplot(x="size" , y="tip" , data=tips)
plt.show()
```

```python
sns.regplot(x="size" , y="tip" , data=tips , x_jitter=0.1)
plt.show()
```

```python
sns.regplot(x="size" , y="tip" , data=tips , x_estimator=np.mean)
plt.show()
```

```python
sns.lmplot(x="total_bill" , y="tip" , data=tips , hue="smoker" )
plt.show()
```

```python
sns.lmplot(x="total_bill" , y="tip" , data=tips ,col="smoker", hue="smoker" )
plt.show()
```

```python
sns.lmplot(x="total_bill" , y="tip" , data=tips ,col="smoker", row="sex", hue="smoker" )
plt.show()
```

```python
tips['smoker']=tips["smoker"]=='Yes'
tips
```

```python
sns.regplot(x="total_bill" , y="smoker", data=tips)
```

## Distribution Plots

```python
penguins=sns.load_dataset("penguins")
penguins
```

```python
sns.histplot(penguins , x="flipper_length_mm")
plt.show()
```

```python
sns.histplot(penguins , x="flipper_length_mm" , bins=20)
plt.show()
```

```python
sns.histplot(penguins , x="flipper_length_mm" , stat="density")
plt.show()
```

```python
sns.histplot(penguins , x="flipper_length_mm" , hue="species")
plt.show()
```

```python
sns.histplot(penguins , x="flipper_length_mm" , kde=True)
plt.show()
```
