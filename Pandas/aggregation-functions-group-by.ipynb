## Aggregate

```python
rng=np.random.RandomState(0)
df=pd.DataFrame({'Key':['A','B','C','A','B','C'],
                'data1':range(6),
                'data2':rng.randint(0,10,6)},
                columns=['Key','data1','data2'])

df
```

Aggregate function apply (min, max , .. other aggregation )functions on specified parts of data 

```python
df.groupby('Key').aggregate([np.min,np.mean])
```

```python
df.groupby('Key').aggregate({'data1':np.min,
                             'data2':np.mean})
```

## filter

The filter() function is used to subset rows or columns of dataframe according to labels in the specified index.

```python
gp=df.groupby('Key')
for(key , group) in gp:
    print(key)
    print(group)
```

```python
def filter_fun(dataframe): # will work in each dataframe or each group
    return dataframe['data2'].mean() < 5
```

```python
df.groupby('Key').filter(filter_fun) # group c has been deleted as mean >=5 
```

## transform

The transform() method allows you to execute a function for each value of the DataFrame

```python
def trans_func(dataframe):
    return dataframe-dataframe.mean()
```

```python
df.groupby('Key').transform(trans_func) # works on column or series
```

## apply

apply() function is used to apply a function along an axis of the DataFrame.

```python
def apply_func(dataframe):
    dataframe['data1']=dataframe['data1']/dataframe['data1'].sum()
    return dataframe
```

```python
df.groupby('Key').apply(apply_func) # works on Dataframes 
```
