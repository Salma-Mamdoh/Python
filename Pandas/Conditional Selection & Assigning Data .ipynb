# Conditional Selection

```python
df=pd.read_csv('ramen-ratings.csv')
```

```python
df.head()
```

To explore data , ask intersting questions 

Explore ramen dishes served in Taiwan

```python
df.loc[df['Country'] =='Taiwan'] # Select all rows when country == Taiwan
```

Have a look at the raman dishes which have rating above than average

```python
# When trying to get the stars more than 2.5 as avg this give an error as data 
# the stars column is str not float as there are some values 'unrates'
```

```python
df=df.loc[df.Stars != 'Unrated'] # to delete all rows which stars value==unrated
```

```python
df.Stars.astype(float) # change type of data to float
```

```python
df.Stars
```

```python
df.loc[df.Stars > 2.5]
```

Served dishes in Taiwan with above average ratings

```python
# & and  , | or , ~ not 
df.loc[(df['Country'] =='Taiwan') & (df['Stars'] > 2.5)]
```

```python
df.loc[(df['Country'] =='Taiwan') & (df['Stars'] > 2.5) , ['Brand','Style']]
```

Have a look at the dishes served in both Finland and sweden

```python
df.loc[(df['Country']=='Finland') | (df['Country']=='Sweden')]
```

```python
df.loc[df.Country.isin(['Finland','Sweden'])]
```

Get the Top Ten values that are not null

```python
df.loc[~df['Top Ten'].isnull()]
```

```python
df.loc[df['Top Ten'].notnull()]
```
