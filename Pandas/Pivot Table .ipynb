```python
import pandas as pd
import numpy as np 
import seaborn as sns
```

```python
taitanic=sns.load_dataset('titanic')
```

```python
taitanic
```

```python
taitanic['sex'].value_counts()
```

```python
taitanic['who'].value_counts()
```

```python
taitanic.pivot_table(index='who' ,columns='class' , values='age' , aggfunc='mean' )
```

```python
taitanic.pivot_table(index='sex' , values='survived' , aggfunc='mean' ,columns='class')
```

```python
taitanic.pivot_table(index='who' , values=['survived','age'] , aggfunc={'survived':'count' , 'age':'mean'} ,columns='class')
```

```python
taitanic.pivot_table(index=['who','sex'] , values=['survived','age'] , aggfunc={'survived':'count' , 'age':'mean'} ,columns='class')
```
